```python
face_cascade = cv2.CascadeClassifier("C:\\Users\\user\\Downloads\\project\\haarcascade_frontalface_default.xml")
```

```python
def face_detect(img):
    faces = face_cascade.detectMultiScale(img)
    for (x,y,w,h) in faces:
        print(x,y,w,h)
        img = cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),3)
    plt.imshow(img)
    plt.show()
```
